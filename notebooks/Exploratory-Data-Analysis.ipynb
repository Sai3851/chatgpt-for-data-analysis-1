# Exploratory Data Analysis (EDA)
This notebook explores the cleaned job postings dataset to identify trends, visualize important features, and gather insights into job locations, job titles, and salaries.


```python
import pandas as pd

# Go up one directory and then into the data folder
df = pd.read_csv('../data/cleaned_gsearch_jobs.csv')

# Display the first few rows to confirm loading
df.head()
```

This step provides the basic information of the dataset along with the basic statistics of the cleaned dataset which is being used for this EDA.

```python
# Check the basic information of the dataset
df.info()

# Display basic statistics for numerical columns
df.describe()
```

As a final step before commencing exploratory data analysis, the dataset is checked for missing values which are already dropped from previous 
standardisation.

```python
# Check for any remaining missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

Null values have been addressed in previous data cleaning steps, particularly for the yearly salary column, allowing us to focus on the most relevant aspects of the dataset for further insights. In this project, the Exploratory Data Analysis (EDA) begins with frequency analysis on three key columns: Job Locations, Job Titles, and Job Platforms. These columns are pivotal as they form the primary inputs for most of the analysis conducted. 

```python
# Top 10 locations by job postings
top_locations = df['location'].value_counts().head(10)
top_locations
```

The frequency analysis of job locations reveals the most common locations for job postings in this dataset.
The top location is "Anywhere," indicating a significant number of remote or location-independent job postings. 
This is followed by general locations like "United States," which may represent jobs that are available nationally or have unspecified locations.

```python
# Top 10 job titles
top_job_titles = df['title'].value_counts().head(10)
top_job_titles
```

The frequency analysis of job titles highlights the most common roles in this dataset.The most frequent role is "Data Analyst," indicating strong demand for general data analysis skills. Variants like "Sr. Data Analyst, Marketing Operations" and "Senior Data Analyst" show a demand for senior-level roles with specific expertise areas, such as marketing operations and business optimization. Other roles, such as "Data Scientist" and "Data Engineer," also appear, reflecting a broader need for data-focused roles.

```python
# Frequency of job posting platforms
platform_counts = df['via'].value_counts()
platform_counts
```

Initially, to understand the top locations in the dataset, a bar chart is drawn using the number of postings listed for each location, based on this the following bar chart is generated.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x=top_locations.values, y=top_locations.index, palette='Blues_r')
plt.title('Top 10 Job Locations')
plt.xlabel('Number of Postings')
plt.ylabel('Location')
plt.tight_layout()
plt.show()
```

Results and Insights:
The "Anywhere" location dominates the chart, suggesting a significant number of remote or location-flexible job postings.
The United States also ranks highly, showing a large number of postings that are either based across multiple regions or have unspecified cities within the country.
Specific cities like Oklahoma City, OK, Kansas City, MO, and Tulsa, OK have notable concentrations of job postings.
There are several smaller locations, including Jefferson City, MO, Bentonville, AR, Wichita, KS, and Dearing, KS, which still make it into the top 10, possibly reflecting regional demand in those areas.

```python
#Based on frequency of distribution of yearly salary, Histogram is drawn
plt.figure(figsize=(10, 6))
sns.histplot(df['salary_yearly'], bins=20, color='skyblue', kde=True)
plt.title('Distribution of Yearly Salaries')
plt.xlabel('Yearly Salary')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

Description:
This histogram displays the distribution of yearly salaries from the job postings dataset. The salary data is plotted on the x-axis, and the frequency of jobs corresponding to each salary range is plotted on the y-axis. A kernel density estimation (KDE) line is overlaid to smooth the distribution and provide insights into the salary trends.

Key Insights:
The peak of the salary distribution is around $100,000, suggesting that this salary range has the highest concentration of job postings.
The distribution is right-skewed, with most salaries clustering between $50,000 and $150,000. There are fewer job postings as salaries increase beyond this range.
A few outliers are visible at higher salary ranges, extending up to $400,000 or more, although these are relatively rare.
The KDE line highlights a single peak, confirming that most salaries are concentrated around the $100,000 range, with a gradual tapering off at higher salaries.
Interpretation:
This salary distribution suggests that the majority of job postings in the dataset offer salaries in the range of $50,000 to $150,000, with the most common salary around $100,000.
The presence of some high-end salaries indicates that there are niche or executive roles with higher compensation, though these are less frequent.
The smooth KDE line provides an indication that the salary data follows a somewhat normal distribution but with a right skew due to the presence of higher salary outliers.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Defining a threshold for aggregating small categories
threshold = 0.02  # Categories with less than 2% will be grouped into "Others"

# Creating a new series with small categories grouped into "Others"
platform_counts_filtered = platform_counts[platform_counts / platform_counts.sum() > threshold]
platform_counts_filtered['Others'] = platform_counts[platform_counts / platform_counts.sum() <= threshold].sum()

# Plotting the pie chart
plt.figure(figsize=(10, 10))  # Increasing the figure size for better clarity
platform_counts_filtered.plot(
    kind='pie',
    autopct='%1.0f%%',  # Showing whole percentages
    startangle=140,
    colors=sns.color_palette('bright'),
    wedgeprops={'edgecolor': 'black'},  # Adding borders to wedges for clarity
    legend=True  # Adding legend to avoid overcrowding with labels
)

plt.title('Distribution of Job Posting Platforms', fontsize=14)
plt.ylabel('')  # Removing the y-axis label
plt.show()

```

Description:
This pie chart shows the distribution of job postings across different platforms. The chart segments represent various job platforms, with each slice indicating the percentage of total job postings attributed to that platform. Platforms with smaller percentages have been grouped under "Others" for clarity.

Key Insights:
LinkedIn dominates the job posting platforms, contributing to 30% of the total postings. This suggests LinkedIn is a major platform for job listings in this dataset.
ZipRecruiter and Indeed follow with 14% and 13%, respectively. These platforms are also popular choices for job postings, showing that a significant portion of jobs is advertised through them.
Bebee and AI Jobs both account for 9% each, indicating that these platforms also have a considerable share of job postings, but less compared to the larger platforms.
Platforms like Built In, Ladders, and several others make up smaller portions, with 3-6% each.
The "Others" category, which aggregates smaller platforms (each contributing less than 2%), constitutes about 15% of the total postings. This suggests that while major platforms dominate, there are many smaller platforms that also contribute to job listings.
Interpretation:
The chart highlights the importance of major job platforms like LinkedIn, ZipRecruiter, and Indeed in the job market, with LinkedIn leading by a significant margin.
The presence of niche platforms such as Bebee and AI Jobs shows that specialized platforms are also utilized for certain roles, particularly in technology and data-focused fields.
The "Others" category signifies a long tail of smaller platforms that collectively account for a notable share of the postings, suggesting that companies utilize a diverse range of platforms beyond the major ones.
